```python
df=pd.read_csv('SampleSuperstore.csv')
```

## Running data Analysis

```python
df.iplot(kind='bar',x='Ship Mode',y='Sales',xTitle='Ship Modes',
         yTitle='Sales in average',title='Sales vs Ship Mode',colors=['blue'],theme='solar')
```

### Observation
Most sales are done through the standard class shipments while least were the number of people who 
bought them the same day

```python
plt.figure(figsize=(15,8))
sns.set_style('whitegrid')
sns.countplot(x='Region',data=df,hue='Segment')
```

### Observation
1)Most Sales are done in the west region of the country. 
2)And most buys are of the consumers and least people have bought it for the home office 

```python
plt.figure(figsize=(20,10))
sns.set_style('whitegrid')
chart=sns.barplot(x='State',y='Sales',data=df)
chart.set_xticklabels(chart.get_xticklabels(),rotation=60)
```

### Observation 
1) Most sales are coming from the state of Wyoming , while the least are coming from the state of South dakota

```python
plt.figure(figsize=(20,10))
sns.set_style('whitegrid')
chart=sns.barplot(x='State',y='Profit',data=df,)
chart.set_xticklabels(chart.get_xticklabels(),rotation=60)
```

### Observation
1) Most profit is coming from the state of Vermont while the least is coming from Columbia

```python
plt.figure(figsize=(15,8))
sns.set_style('whitegrid')
sns.barplot(x='Category',y='Sales',data=df,hue='Segment')
```
